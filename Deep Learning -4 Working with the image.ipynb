```python
# Affine Transformation
import tensorflow as tf
transformation = tf.keras.preprocessing.image.apply_affine_transform(img,theta=30)
plt.imshow(transformation)
```

```python
# zx: Zoom in x direction.
# zy: Zoom in y direction

transformation = tf.keras.preprocessing.image.apply_affine_transform(img,zx=0.5,zy= 0.5)
plt.imshow(transformation)
```

```python
# tx: Width shift.
# ty: Height shift.

transformation = tf.keras.preprocessing.image.apply_affine_transform(img,tx=50,ty=100)
plt.imshow(transformation)
```
