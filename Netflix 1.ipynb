### Preparing the dataset for the movies of the 2016

```python
import pandas as pd
import numpy as np

data=pd.read_csv("E:/Full Stack Data Scientist Bootcamp/Recommandation System/rc/archive (1)/movie_metadata.csv")
data
```

```python
df=data[['director_name','actor_1_name','actor_2_name','actor_3_name','genres','movie_title']]
```

```python
df
```

```python
df['movie_title']=df['movie_title'].str.lower()
df
```
