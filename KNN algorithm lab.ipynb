### EDA (Exploratory Data Analysis)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df=pd.read_csv("E:\Full Stack Data Scientist Bootcamp\Supervised ML resources/wisc_bc_data.csv")
```

```python
df.head()
```

```python
df.dtypes
```

```python
df['diagnosis']=df['diagnosis'].astype('category')
```

```python
df.dtypes
```

```python
df=df.drop('id',axis=1)
```

```python
#Now divides the into independant and dependant variable
y=df['diagnosis']
x=df.drop('diagnosis',axis =1,)
```

```python
x
```

```python
# for the scaling of the data frist the above step (seprating the dependent and independent features of data) 
# should be done so that we can avoid from the Data Leak

from scipy.stats import zscore
Xscaled=x.apply(zscore)
Xscaled
```

```python
x=pd.DataFrame(Xscaled)
x
```

```python
x.describe()
```

### Building the Model of KNN 

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.30,random_state=0)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier(n_neighbors=5,weights='distance')
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
from sklearn import metrics
cm=metrics.confusion_matrix(y_test,y_pred,labels=['M','B'])

# df_cm=pd.DataFrame(cm, index=[i for i in ["M","B"]], columns=[i for i in ["Predict M",'predict B']])

plt.figure(figsize=(7,5))
# sb.heatmap(df_cm,annot=True)
sb.heatmap(cm,annot=True)
```

```python
model.score(x_test,y_test)
```

### Choosing the K-values

```python
from sklearn.model_selection import cross_val_score
scores_1=[]

for i in range(1,50):
    model_2=KNeighborsClassifier(n_neighbors=i)
    score_2=cross_val_score(model_2,x,y,cv=10)
    scores_1.append(score_2.mean())
```

```python
score_2
```

```python
len(scores_1)
```

```python
plt.figure(figsize=(20,10))
plt.title('Accuracy rate Vs K-Values')
plt.plot(range(1,50),scores_1,color='blue',linestyle='dashed',marker='o',markerfacecolor='red',markersize=10)

plt.xlabel('k')
plt.ylabel('Accuracy')
plt.show()
```
