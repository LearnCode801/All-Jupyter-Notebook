#### 3. Target Guided Ordinal Encoding
###### i) Ordering the according to the target
###### ii) Replace the labels by the Probability of being 1 or 0

```python
df=pd.read_csv('E:Full Stack Data Scientist Bootcamp/datasets/titanic.csv', usecols=['Survived','Cabin'])
df
```

```python
df.Cabin.fillna('Missing',inplace=True)
```

```python
df
```

```python
df['Cabin']=df['Cabin'].astype(str).str[0]
```

```python
df.Cabin.unique()
```

```python
df.groupby(['Cabin'])['Survived'].mean()
```

```python
lables=df.groupby(['Cabin'])['Survived'].mean().sort_values().index
lables
```

```python
a=0;
dict_rank={}
for i in lables:
    dict_rank[i]=a
    a=a+1
    
```

```python
dict_rank
```

```python
df['new_by_the_target_guided_Ordinal_encoding']=df['Cabin'].map(dict_rank)
```

```python
df
```

#### Target- Guaided Mean Encoding

```python
sort_category_wise_mean.to_dict()
```

#### Probability Rtio Encoding 

```python
df=pd.read_csv('E:Full Stack Data Scientist Bootcamp/datasets/titanic.csv', usecols=['Survived','Cabin'])
df
```

```python
df.Cabin.fillna('Missing',inplace=True)
```

```python
df['Cabin']=df['Cabin'].astype(str).str[0]
```

```python
probalities_cabin_on_servived=df.groupby(['Cabin'])['Survived'].mean().sort_values()
probalities_cabin_on_servived
```

```python
table=pd.DataFrame(probalities_cabin_on_servived)
```

```python
table
```

```python
table['Dided Probabilities']=1-table['Survived']
```

```python
table
```

```python
table['ratio_of_live_dead']=table['Survived']/table['Dided Probabilities']
```

```python
table
```

```python
dcit=table['ratio_of_live_dead'].to_dict()
```

```python
dcit
```

```python
df['ratio_encoding']=df['Cabin'].map(dcit)
```
