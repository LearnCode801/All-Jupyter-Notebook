# Linear Regression LAB-2(part-2)

```python
# Linear Regression LAB-2(part-2)
# starting the making model 
y=cData['mpg']
y
```

```python
x=cData.drop('mpg',axis=1)
x
```

```python
# spliting the data
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.30,random_state=0)
```

```python
y_test
```

```python
#Making the object of LinearRegression class
model=LinearRegression()
```

```python
#Train the model by the part of test-data  with the original (input and output) data 
model.fit(x_train,y_train)
```

```python
model.coef_
```

```python
for idx, col_name in enumerate(x_train.columns):
    print('the coefficiant for {} if {}'.format(col_name,model.coef_[idx]))
```

```python
model.intercept_
```

```python
from sklearn import metrics
prediction_of_test_mpg = model.predict(x_test)
MAE=metrics.mean_absolute_error(y_test,prediction_of_test_mpg)
MAE
```

```python
RMSE=np.sqrt(MAE)
RMSE
```
