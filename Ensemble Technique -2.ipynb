```python
replaceData={
            "checking_balance":     {"< 0 DM": 1, "1 - 200 DM": 2 ,"> 200 DM": 3 ,"unknown":-1},
            "credit_history": {"critical": 1, "poor":2 , "good": 3, "very good": 4,"perfect": 5},
            "savings_balance": {"< 100 DM": 1, "100 - 500 DM":2 , "500 - 1000 DM": 3, "> 1000 DM": 4,"unknown": -1},
            "employment_duration":     {"unemployed": 1, "< 1 year": 2 ,"1 - 4 years": 3 ,"4 - 7 years": 4 ,"> 7 years": 5},
            "phone":     {"no": 1, "yes": 2 },
            "default":     {"no": 0, "yes": 1 }
}
oneHotCols=["purpose","housing","other_credit","job"]
```

```python
creditData=creditData.replace(replaceData)
```

```python
creditData.tail()
```

```python
creditData.dtypes
```

```python
creditData=pd.get_dummies(creditData, columns=oneHotCols)
```

```python
creditData.tail()
```

```python
creditData.dtypes
```

```python
y=creditData['default']
x=creditData.drop('default',axis=1)
```

```python
x
```

```python
y
```

```python
x_train, x_test, y_train ,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
model_1= DecisionTreeClassifier(criterion='gini' ,random_state=1)
model_1.fit(x_train,y_train)
```

```python
y_pred=model_1.predict(x_test,y_test)
```

```python
print(model_1.score(x_train,y_train))
print(model_1.score(x_test,y_test))
```

```python
train_char_label = ['No', 'Yes']
Credit_Tree_File = open('credit_tree.dot','w')
dot_data = tree.export_graphviz(model_1, out_file=Credit_Tree_File, feature_names = list(x_train), class_names = list(train_char_label))
Credit_Tree_File.close()
```
