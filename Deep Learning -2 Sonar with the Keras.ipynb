```python
import numpy as np
import pandas as pd
import keras
```

```python
df=pd.read_csv('E:/Full Stack Data Scientist Bootcamp/Deep Learning/datasets/sonar.csv',delimiter=',')
df
```

```python
x=df.drop('Class',axis=1)
y=df.Class
```

```python
x
```

```python
y
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y=le.fit_transform(y)
print (y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=1)

_, x_val,_, y_val = train_test_split(x_test,y_test, test_size=0.25, random_state=1) 
```

```python
_
```

```python
# building the NN model
import tensorflow as tf
from tensorflow import keras
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = keras.Sequential([
    keras.layers.Dense(80, input_dim=60, activation='relu'),
    keras.layers.Dense(30, activation='relu'),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

H = model.fit(x_train, y_train, epochs=100, batch_size=8, validation_split=0.3)
```

```python
loss,accuracy=model.evaluate(x_train,y_train)
print(' Accuracy {} \n Loss {}'.format(accuracy,loss))
```

```python
loss,accuracy=model.evaluate(x_test,y_test)
print(' Accuracy {} \n Loss {}'.format(accuracy,loss))
```

```python
# model is overfitted
H.history
```

```python
H.history['loss']
```

```python
import matplotlib.pyplot as plt


epochs = np.arange(0, len(H.history["loss"]))
plt.figure(figsize=(10,8))
plt.style.use("ggplot")


# plot training and validation loss
plt.style.use("ggplot")
plt.plot(epochs, H.history["loss"], label="train_loss")
plt.plot(epochs, H.history["val_loss"], label="val_loss")
plt.title("Training Vs Validation Loss")
plt.xlabel("Epoch #")
plt.ylabel("Loss")
plt.legend()
```

```python
# Removing the Overfitting 

# Simplify the complexity of the model
# L2 Regularization
# Drop Out
# Early Stopping
# Data Augmentation
```

```python
# L2 Regularization
#kernel_regularizer=tf.keras.regularizers.l2(0.1)

model = keras.Sequential([
    keras.layers.Dense(80, input_dim=60, activation='relu',kernel_regularizer=tf.keras.regularizers.l2(1)),
    keras.layers.Dense(30, activation='relu'),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

H = model.fit(x_train, y_train, epochs=100, batch_size=8, validation_split=0.3)
```

```python
loss,accuracy=model.evaluate(x_train,y_train)
print(' Accuracy {} \n Loss {}'.format(accuracy,loss))

loss,accuracy=model.evaluate(x_test,y_test)
print(' Accuracy {} \n Loss {}'.format(accuracy,loss))
```

```python
plt.style.use("ggplot")
plt.plot(epochs, H.history["loss"], label="train_loss")
plt.plot(epochs, H.history["val_loss"], label="val_loss")
plt.title("Training Vs Validation Loss")
plt.xlabel("Epoch #")
plt.ylabel("Loss")
plt.legend()
```

```python
# Drop Out
# keras.layers.Dropout(0.1),

model = keras.Sequential([
    keras.layers.Dense(80, input_dim=60, activation='relu'),
    keras.layers.Dropout(0.1),
    keras.layers.Dense(30, activation='relu'),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

H = model.fit(x_train, y_train, epochs=100, batch_size=8, validation_split=0.3)
```

```python
loss,accuracy=model.evaluate(x_train,y_train)
print(' Accuracy {} \n Loss {}'.format(accuracy,loss))

loss,accuracy=model.evaluate(x_test,y_test)
print(' Accuracy {} \n Loss {}'.format(accuracy,loss))
```

```python
plt.style.use("ggplot")
plt.plot(epochs, H.history["loss"], label="train_loss")
plt.plot(epochs, H.history["val_loss"], label="val_loss")
plt.title("Training Vs Validation Loss")
plt.xlabel("Epoch #")
plt.ylabel("Loss")
plt.legend()
```

```python
# Removing the overfitting of model by using the both Drop Out and L2 Regularization
# 1. kernel_regularizer=tf.keras.regularizers.l2(1)
# 2. keras.layers.Dropout(0.1)

model = keras.Sequential([
    keras.layers.Dense(80, input_dim=60, activation='relu',kernel_regularizer=tf.keras.regularizers.l2(1)),
    keras.layers.Dropout(0.1),
    keras.layers.Dense(30, activation='relu'),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

H = model.fit(x_train, y_train, epochs=100, batch_size=8, validation_split=0.3)
```

```python
loss,accuracy=model.evaluate(x_train,y_train)
print(' Accuracy {} \n Loss {}'.format(accuracy,loss))

loss,accuracy=model.evaluate(x_test,y_test)
print(' Accuracy {} \n Loss {}'.format(accuracy,loss))
```

```python
plt.style.use("ggplot")
plt.plot(epochs, H.history["loss"], label="train_loss")
plt.plot(epochs, H.history["val_loss"], label="val_loss")
plt.title("Training Vs Validation Loss")
plt.xlabel("Epoch #")
plt.ylabel("Loss")
plt.legend()
```

```python
# Early Stoping

model = keras.Sequential([
keras.layers.Dense(80, input_dim=60, activation='relu'),
keras.layers.Dense(30, activation='relu'),
keras.layers.Dense(15, activation='relu'),
keras.layers.Dense(1, activation='sigmoid')
])

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

earlystopping = tf.keras.callbacks.EarlyStopping(monitor='val_loss',verbose=1, patience=5)

callbacks = [earlystopping]


H = model.fit(x_train, y_train, epochs=100, batch_size=8, validation_split=0.3,callbacks=callbacks) #To discover the training epoch on which training was stopped, the “verbose” argument can be set to 1. Once stopped, the callback will print the epoch number., Add "patience" to wait for sometime before stopping after the best accuracy or val_loss has been found.
```

```python
loss,accuracy=model.evaluate(x_train,y_train)
print(' Accuracy {} \n Loss {}'.format(accuracy,loss))

loss,accuracy=model.evaluate(x_test,y_test)
print(' Accuracy {} \n Loss {}'.format(accuracy,loss))
```

```python
# ModelCheckpoint Callback Technique


from tensorflow.keras.callbacks import ModelCheckpoint

model = keras.Sequential([
keras.layers.Dense(80, input_dim=60, activation='relu'),
keras.layers.Dense(30, activation='relu'),
keras.layers.Dense(15, activation='relu'),
keras.layers.Dense(1, activation='sigmoid')
])

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

checkpoint = ModelCheckpoint('SOMETHINGELSE', monitor="val_loss", mode="min",save_best_only=True, verbose=1)


H = model.fit(x_train, y_train, epochs=100, batch_size=8, validation_split=0.3,callbacks=checkpoint) #To discover the training epoch on which training was stopped, the “verbose” argument can be set to 1. Once stopped, the callback will print the epoch number., Add "patience" to wait for sometime before stopping after the best accuracy or val_loss has been found.
```

```python
loss,accuracy=model.evaluate(x_train,y_train)
print(' Accuracy {} \n Loss {}'.format(accuracy,loss))

loss,accuracy=model.evaluate(x_test,y_test)
print(' Accuracy {} \n Loss {}'.format(accuracy,loss))
```

```python
# Combining the all Techinque

from tensorflow.keras.callbacks import ModelCheckpoint

model = keras.Sequential([
keras.layers.Dense(80, input_dim=60, activation='relu',kernel_regularizer=tf.keras.regularizers.l2(1)),
keras.layers.Dropout(0.1),
keras.layers.Dense(30, activation='relu'),
keras.layers.Dense(15, activation='relu'),
keras.layers.Dense(1, activation='sigmoid')
])

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

earlystopping = tf.keras.callbacks.EarlyStopping(monitor='val_loss',verbose=1, patience=5)
checkpoint = ModelCheckpoint('SOMETHINGELSE', monitor="val_loss", mode="min",save_best_only=True, verbose=1)


callbacks = [earlystopping,callbacks]


H = model.fit(x_train, y_train, epochs=100, batch_size=8, validation_split=0.3,callbacks=callbacks) #To discover the training epoch on which training was stopped, the “verbose” argument can be set to 1. Once stopped, the callback will print the epoch number., Add "patience" to wait for sometime before stopping after the best accuracy or val_loss has been found.
```

```python
loss,accuracy=model.evaluate(x_train,y_train)
print(' Accuracy {} \n Loss {}'.format(accuracy,loss))

loss,accuracy=model.evaluate(x_test,y_test)
print(' Accuracy {} \n Loss {}'.format(accuracy,loss))
```
