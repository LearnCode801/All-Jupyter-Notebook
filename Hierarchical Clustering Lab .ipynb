#### Making the Agglomerative Clustring 

```python
from sklearn.cluster import AgglomerativeClustering 
model = AgglomerativeClustering(n_clusters=3, affinity='euclidean',linkage='average')
```

```python
model.fit(data_2)
```

```python
model.labels_
```

```python
prediction=model.predict(data_2)
```

```python
data['labels']=model.labels_
data
```

#### Making the dendrograme

```python
# making the dendrogram
from scipy.cluster.hierarchy import cophenet, dendrogram , linkage
```

```python
from scipy.spatial.distance import pdist

z=linkage(data_2,metric='euclidean',method='average')
```

```python
c, cophenet = cophenet (z ,pdist (data_2))
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
plt.xlabel('simle index')
plt.ylabel('Distance')
plt.title("Agglomerative Hierechical Clustering ")
dendrogram(z,leaf_rotation=30,leaf_font_size=10)
plt.tight_layout()
```
