```python
import numpy as np
from numpy import loadtxt 
import keras 
from keras.models import Sequential
from keras.layers import Dense
```

```python
dataset=loadtxt('E:/Full Stack Data Scientist Bootcamp/datasets/datasets-20221208T145556Z-001/datasets/pima-indians-diabetes.csv',delimiter=',',)
```

```python
print(dataset)
```

```python
x=dataset[:,0:8]
y=dataset[:,8:]
print(x)
print(y)
```

```python
import pandas as pd
x_df=pd.DataFrame(x)
y_df=pd.DataFrame(y)
```

```python
x_df.head()
```

```python
y_df.head()
```

```python
model = Sequential()
```

```python
model.add(Dense(units=12,input_dim=8,activation='relu'))
model.add(Dense(units=9,activation='relu'))
model.add(Dense(units=1,activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(x,y,epochs=150,batch_size=10)
```

```python
loss,accuracy=model.evaluate(x,y)
print(' Accuracy {} \n Loss {}'.format(accuracy,loss))
```

```python
prediction=model.predict(x)
prediction
```
