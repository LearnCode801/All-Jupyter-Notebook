```python
import requests
import bs4
import pandas as pd
```

```python
ourdata=requests.get('https://en.wikipedia.org/wiki/Artificial_intelligence')
ourdata.text
```

```python
ourdata
```

```python
bsp=bs4.BeautifulSoup(ourdata.text,'xml')
bsp
```

```python
titles=bsp.select('h2')
titles
```

```python
titles_class=bsp.select('h3')
titles_class
```

```python
for i in titles_class:
    print (i.text)
```

```python
images=bsp.select('img')
images
```
